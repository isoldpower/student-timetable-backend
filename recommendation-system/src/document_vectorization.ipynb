```python
import pandas as pd
import numpy as np
import gensim
```

```python
df = pd.read_parquet('../data/courses_preprocessed.parquet')
```

```python
df
```

```python
model = gensim.models.KeyedVectors.load_word2vec_format('../data/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
documents = df['token_desc'].to_list()
```

### Averaging word embeddings to convert to document embeddings

```python
doc_vec = []

for i in range(len(documents)):
    sent_doc_vec = np.array(100)
    for j in range(len(documents[i])):
        sent_doc_vec = np.add(sent_doc_vec, model[documents[i][j]])

    sent_doc_vec = np.divide(sent_doc_vec, len(documents[i]))
    doc_vec.append(sent_doc_vec)
```

```python
embed_df = pd.DataFrame(index=df.course_id, data=doc_vec)
```

```python
embed_df
```

```python
embed_df.columns = embed_df.columns.astype(str)
embed_df.to_parquet('../data/doc_vec_embeddings.parquet')
```
